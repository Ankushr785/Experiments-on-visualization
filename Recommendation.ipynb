```python
import operator

def Pearson_Correlation():
    data = np.array(df.values)
    rows = df.index.values
    
    (nrows,ncols) = data.shape
    
    Corr_val = {}
     
    for row1 in range(nrows):
        for row2 in range(nrows):
            if row1 != row2 :
                data_1 = [data[row1, :][n] for n in range(ncols) if data[row1,n]!=0 and data[row2,n] != 0]    
                data_2 = [data[row2, :][n] for n in range(ncols) if data[row1,n]!=0 and data[row2,n] != 0]    

                data_1 = np.array(data_1) - np.mean(data_1)
                data_2 = np.array(data_2) - np.mean(data_2)

                p_corr  = np.dot(data_1,data_2)*1.0 / np.sqrt(sum(np.power(data_1,2))*sum(np.power(data_2,2)))

                Corr_val[(rows[row1],rows[row2])] = p_corr  
            
            
            
    return Corr_val        
            
            
def Correlation_rank(user_id,n):
    
    p_corr = Pearson_Correlation()
    p_corr_row  = {}
    
    for i, j in p_corr.keys():
        if i == user_id:
            p_corr_row[(i,j)] = p_corr[(i,j)]

    sorted_p = sorted(p_corr_row,reverse = True, key=operator.itemgetter(1))[:n]         
    #print (p_corr_row)
    new_return = {} 
    for i , j in sorted_p:
        new_return[(i,j)] = p_corr_row[(i,j)]
        
    print (sorted_p)        
    return  new_return        



def prediction(user_id , n):
    
    ranked_p_rel = Correlation_rank(user_id,n)
    neigh_users = [b for(a,b) in  ranked_p_rel.keys()]
    
    #weights = ranked_p_rel
    
    pred = []
    
    
    
    for col in df.columns:
        vals = 0
        wghts = 0
        
        for neighb in neigh_users:
            val = df.loc[neighb,col]
            print val
            if val == 0:
                wght = 0
            
            
            else:
                wght = ranked_p_rel[(user_id,neighb)]
                
            vals += wght*val
            wghts += wght
            
            if wghts == 0:
                pred.append(0)
                
            else:
                pred.append(vals*1.0 / wghts)
                
                
    return pred



def recom(user_id,n,m):
    cols = df.columns.values
    pred = prediction(user_id,n)
    
    pred_dict={}
    #print pred
    for i in range(len(cols)):
        pred_dict[cols[i]] = pred[i]
        
    
    sorted_p = sorted(pred_dict,reverse = True, key=operator.itemgetter(1))[:m] 
    
   
    #print (sorted_p)        
    #return  new_return  
     
    for item in sorted_p:
        print item, "  ",pred_dict[item]
        
        
        
 



recom(user_id = 9,n = 4, m = 2)


            
              
                
    
    
    
    
    
    
    
    

            
            
            
            
        
        
    
    



```

```python
# we have user_table which will give rating to a particular dish
# content based recommeder system
# we have 2 tables with us

#one is food - spices table and another is food-user rating table
#we will use tf-idf approach




#matrix of spice will contain only of 1 and 0's so need to normalize it

#normalization along the row
def spice_norm():
    
    num_attr = np.sum(spice,axis=1)
    doc_vec = []
    for i in range(num_attr.shape[0]):
        doc_vec.append(np.sqrt(spice[i,:]*1.0/num_attr[i]))
            
    return np.array(doc_vec)      
    
    
    
def attr_norm():
    
    num_attr = np.sum(spice,axis=0)
    
    doc_vec = []
    for i in range(num_attr.shape[0]):
        doc_vec.append(spice[:,i]*1.0 / num_attr[i])

    return np.array(doc_vec)

def user_v(spice_mat):
    user_vec = []
    
    for i in range(user.shape[1]):
        user_vec.append([np.dot(user[:,i],spice_mat[:,j]) for j in range(spice_mat.shape[1])])
    
    return np.array(user_vec)
    
def weight_predict(spice_mat):
        
    user_vec = user_v(spice_mat)
    
    preds = []
    
    for i in range(spice_mat.shape[0]):
        preds.append([np.dot(spice_mat[i,:],user_vec[j,:]) for j in range(user_vec.shape[0])])

    return np.array(preds)
    
    
def idf_weight_pred():
    spice_mat = spice_norm()
    attr_wght = attr_norm()
    print attr_wght
    weighted_spice = []
    
    for i in range(spice_mat.shape[0]):
        weighted_spice.append([np.dot(spice_mat[i,:],attr_wght[:,j]) for j in range(attr_wght.shape[0])])
        
    preds  = weight_predict(spice_mat = np.array(weighted_spice))     

    return np.array(preds)
    


preds = idf_weight_pred()

print preds










```
